```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
x=df.drop(columns=['species'])
y=df['species']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
lr=LogisticRegression()
dt=DecisionTreeClassifier()
kn=KNeighborsClassifier()
```

```python
voting_hard=VotingClassifier(
    estimators=[('Logistic',lr),('DecisionTree',dt),('Kneighbors',kn)],voting='hard')
```

```python
voting_hard.fit(x_train,y_train)
```

```python
y_pred=voting_hard.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('Accuracy_Score of Hard Voting',accuracy_score(y_test,y_pred))
```

```python
voting_hard2=VotingClassifier(
    estimators=[('Logistic',lr),('DecisionTree',dt),('Kneighbors',kn)],voting='soft')
```

```python
voting_hard2.fit(x_train,y_train)
```

```python
y_pred2=voting_hard2.predict(x_test)
```

```python
print('accuracy_score of soft voting',accuracy_score(y_test,y_pred2))
```
